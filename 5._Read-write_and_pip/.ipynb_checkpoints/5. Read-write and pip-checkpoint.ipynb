# Домашнее задание
1. Переведите содержимое файла purchase_log.txt в словарь purchases вида:
```python
{'1840e0b9d4': 'Продукты', ...}
```

2. Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

````python
import csv
import json

with open('purchase_log.txt', 'r') as txt_file:
    purchases_list = []
    for line in txt_file:
        purchase = {}
        for i, item in enumerate(line.strip().replace('{', '').replace('}', '').replace('"', '').split(',')):
            purchase[item.split(':')[0].strip()] = item.split(':')[1].strip()
        purchases_list.append(purchase)

with open('purchase.json', 'w') as json_file:
    json.dump(purchases_list, json_file)
    print(f'Файл "purchase.json" записан.')


with open('visit_log.csv', 'r', newline='') as csv_file:
    fieldnames = ['user_id', 'source']
    fieldnames_new = ['user_id', 'source', 'category']
    with open('funnel.csv', 'w', newline='') as new_csv_file:
        print(f"Сравниваем два массива и результат записываем в 'funnel.csv'.")
        new_file = csv.DictWriter(new_csv_file, fieldnames=fieldnames_new)
        for line in csv.DictReader(csv_file, fieldnames=fieldnames):
            for item in purchases_list:
                if line['user_id'] == item['user_id']:
                    new_file.writerow({
                        'user_id': line['user_id'],
                        'source': line['source'],
                        'category': item['category'],
                        }
                    )
                    print({
                        'user_id': line['user_id'],
                        'source': line['source'],
                        'category': item['category'],
                        }
                    )
        print('Запись завершена.')
````
