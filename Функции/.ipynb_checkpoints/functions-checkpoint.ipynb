Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


class Catalog:
    def __init__(self, docs, dirs):
        self.docs = docs
        self.dirs = dirs
        self.error = 'Вы ввели пустое значение или документ не найден...'

    def is_shelf(self, number, shelf_target):
        return ((str(shelf_target) in list(self.dirs.keys())) and\
                (self.shelf(number, result=True) != str(shelf_target)) and\
                (self.shelf(number, result=True) != self.error))

    def is_number(self, number, shelf_target):
        if len([item for item in list(self.dirs.values()) if (str(number) in item)]) > 0:
            return True
        else:
            return False

    def people(self, number):
        try:
            print(
                f"У документа {number} владелец: {list(result['name'] for result in self.docs if result['number'] == str(number))[0]}")
        except IndexError:
            print(self.error)

    def shelf(self, number, result=False):
        try:
            shelf = [key for key, value in self.dirs.items() if list(shelf for shelf in value if str(number) == shelf)][
                0]
            if result:
                return shelf
            else:
                print(f'У документа {number} полка: {shelf}.')
        except IndexError:
            if result:
                return self.error
            else:
                print(self.error)

    def list(self):
        for row in self.docs:
            print(f"passport: {row['number']} {row['name']}")

    def add_shelf(self, shelf):
        for key, value in self.dirs.items():
            if key == str(shelf):
                return print('Такая полка уже существует...')
        self.dirs[str(shelf)] = []
        print(f'Полка {shelf} добавлена.')

    def delete_from_dirs(self, number):
        try:
            shelf_number = self.shelf(number, result=True)
            dirs_length_before_delete = len(self.dirs[shelf_number])
            for key, value in self.dirs.items():
                directories[key] = [item for item in value if str(number) not in item]
            if len(self.dirs[shelf_number]) < dirs_length_before_delete:
                return print('Документ удален из полок.')
            return print('Документ не найден на полке.')
        except KeyError:
            print(self.error)

    def delete_from_docs(self, number):
        docs_length_before_delete = len(self.docs)
        self.docs = [dic for dic in self.docs if dic.get('number') != number]
        if len(self.docs) < docs_length_before_delete:
            return print('Документ удален из каталога.')
        return print('Документ не найден в каталоге.')

    def move(self, number, shelf_target):
        if self.is_shelf(number, shelf_target):
            for key, value in self.dirs.items():
                self.dirs[key] = [item for item in value if number not in item]
            self.dirs[str(shelf_target)].append(number)
            return print('Документ перемещен.')
        return print('Документ не существует или уже находится на полке.')

    def add(self, number, shelf_target, **kwargs):
        if not self.is_shelf(number, shelf_target) and not self.is_number(number, shelf_target):
            self.docs.append({
                'type': kwargs['type'],
                'number': number,
                'name': kwargs['name'],
            },
            )
            self.dirs[str(shelf_target)].append(number)
            return print('Документ добавлен.')
        return print('Такой документ уже существует.')


class Main:
    def __init__(self):
        self.catalog_object = Catalog(documents, directories)

    def start(self, command, doc_number=None, shelf_number=None, doc_type=None, person_name=None):
        if command == 'p':
            self.catalog_object.people(doc_number)
        elif command == 's':
            self.catalog_object.shelf(doc_number)
        elif command == 'l':
            self.catalog_object.list()
        elif command == 'as':
            self.catalog_object.add_shelf(shelf_number)
        elif command == 'd':
            self.catalog_object.delete_from_docs(doc_number)
            self.catalog_object.delete_from_dirs(doc_number)
        elif command == 'm':
            self.catalog_object.move(doc_number, shelf_number)
        elif command == 'a':
            self.catalog_object.add(doc_number, shelf_number, type=doc_type, name=person_name)
        else:
            print('Неизвестная команда')


if __name__ == "__main__":

    catalog = Main()
    catalog.start('p', doc_number="2207 876234")
    catalog.start('p', doc_number="2207 87623")
    catalog.start('p', doc_number="")
    catalog.start('p', doc_number=0)
    print('******')
    catalog.start('s', doc_number="2207 876234")
    catalog.start('s', doc_number="2207 87623")
    catalog.start('s', doc_number="")
    catalog.start('s', doc_number=0)
    print('******')
    catalog.start('l')
    print('******')
    catalog.start('as', shelf_number=1)
    catalog.start('as', shelf_number=2)
    catalog.start('as', shelf_number=3)
    catalog.start('as', shelf_number=4)
    print('******')
    catalog.start('d', doc_number='2207 876234')
    catalog.start('d', doc_number='11-3')
    print('******')
    catalog.start('m', doc_number='11-2', shelf_number=2)
    catalog.start('m', doc_number='11-2', shelf_number=1)
    catalog.start('m', doc_number='11-3', shelf_number=2)
    catalog.start('m', doc_number='11-2', shelf_number=4)
    print('******')
    catalog.start('a', doc_number='777 777', shelf_number=4, doc_type='passport', person_name='John Smith')
    catalog.start('a', doc_number='11-2', shelf_number=1, doc_type='invoice', person_name='Геннадий Покемонов')
    catalog.start('a', doc_number='11-2', shelf_number=2, doc_type='invoice', person_name='Геннадий Покемонов')
    catalog.start('a', doc_number='11-2', shelf_number=3, doc_type='invoice', person_name='Геннадий Покемонов')
```
